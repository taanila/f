### Analyysejä analysointiosaamisen tehtävään 2

```python
#Kirjastojen tuonti
import pandas as pd
import scipy.stats as stats
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

#Datan lukeminen
df=pd.read_excel('http://myy.haaga-helia.fi/~taaak/a/t2aineisto.xlsx')
df.head()
```

```python
#Lyhyemmät nimet muuttujille
df.columns=['nro', 'nimi', 'tuotto5', 'tuotto3', 'tuotto1', 'tuotto6kk', 'palkkio', 'luokka']
df.head()
```

```python
#Rahastojen lukumäärä eri rahastoluokissa
df['luokka'].value_counts()
```

```python
#Tilastolliset tunnusluvut rahastoluokittain
df.groupby('luokka')[['tuotto5','tuotto3','tuotto1','tuotto6kk','palkkio']].describe().T
```

```python
#5 vuoden tuottojen jakauma
df.boxplot(column='tuotto5', by='luokka')
plt.title('')
plt.suptitle('')
plt.xlabel('Rahastoluokka')
plt.ylabel('5 vuoden tuotto')
```

```python
#3 vuoden tuottojen jakauma
df.boxplot(column='tuotto3', by='luokka')
plt.title('')
plt.suptitle('')
plt.xlabel('Rahastoluokka')
plt.ylabel('3 vuoden tuotto')
```

```python
#Vuoden tuottojen jakauma
df.boxplot(column='tuotto1', by='luokka')
plt.title('')
plt.suptitle('')
plt.xlabel('Rahastoluokka')
plt.ylabel('1 vuoden tuotto')
```

```python
#6 kuukauden tuottojen jakauma
df.boxplot(column='tuotto6kk', by='luokka')
plt.title('')
plt.suptitle('')
plt.xlabel('Rahastoluokka')
plt.ylabel('6 kuukauden tuotto')
```

```python
#Palkkioiden jakauma
df.boxplot(column='palkkio', by='luokka')
plt.title('')
plt.suptitle('')
plt.xlabel('Rahastoluokka')
plt.ylabel('Hallinn. ja säilytyspalkkio')
```

```python
#Data vuoden tuotoista ja palkkioista ilman puuttuvia arvoja
dfdropna=df[['tuotto1','palkkio','luokka']].dropna()
dfdropna.head()
```

```python
#Vuoden tuoton ja palkkion välinen korrelaatiokerroin ja p-arvo
stats.pearsonr(dfdropna['tuotto1'], dfdropna['palkkio'])
```

```python
#Korrelaatiokertoimet ja p-arvot rahastoluokittain
luokat=dfdropna['luokka'].unique()
for value in luokat:
    dfluokka=dfdropna[dfdropna['luokka']==value]
    print(value, stats.pearsonr(dfluokka['tuotto1'], dfluokka['palkkio']))
```

```python
#Korrelaation tarkempi tarkastelu EQ_FINLAND-rahastoluokalle
from scipy.stats import pearsonr
sns.jointplot(x='palkkio', y='tuotto1', data=df[df['luokka']=='EQ_FINLAND'], 
              kind='reg', stat_func=pearsonr)
```

```python
#Muuttujien väliset riippuvuudet hajontakaavioina
sns.pairplot(df[['tuotto5','tuotto3','tuotto1','tuotto6kk','palkkio']].dropna(), kind='reg')
```
